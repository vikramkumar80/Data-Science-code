# filter with More than one Conditions(And)

```python
df = pd.read_csv("employees.csv",parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2].head()
```

# Filter with More than One Condition(OR)

```python
df = pd.read_csv("employees.csv",parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"
df[mask1 & mask2].head(10)
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2)| mask3]
```

# The .isin()  Method

```python
df = pd.read_csv("employees.csv",parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] > "product"

df[mask1 | mask2 | mask3].head(10)
```

#  The   .isnull() and  .notnull() Methods

```python
df = pd.read_csv("employees.csv",parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask = df ["Team"].isnull()

df[mask]
```

```python
condition = df["Team"].notnull()

df[condition].head()
```

# The .between() Method

```python
df = pd.read_csv("employees.csv",parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
df[df["Salary"].between(60000, 70000)].head(10)
```

```python
df[df["Bonus %"].between(2.0 ,5.0)].head(10)
```

```python
df[df["Start Date"].between("1991-01-01", "1992-01-01")].head(10)
```

```python
df[df["Last Login Time"].between("08:30AM","12:00PM")].head(10)
```

# The .duplicated() Method

```python
df = pd.read_csv("employees.csv",parse_dates = ["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```
