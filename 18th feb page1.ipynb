# sort a dataframe with a .sort_value() method

```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv");
```

```python
nba.head(5)
```

```python
nba.sort_values(["Team","Name"],ascending=False).head()
```

```python
nba.sort_values(["Team","Name"],ascending=[True,False]).head(5)
```

```python
nba.sort_values(["Team","Name"],ascending=[False,True]).head(5)
```

```python
nb=nba.sort_values(["Team","Name"],ascending=[True,False]).head(5)
```

```python
nb.head()
```

```python
nba.sort_values(["Team","Name"],ascending=[False,True],inplace=True)
```
